````python
matrix_1 = np.ones((5, 3), dtype= np.int8)
matrix_2 = np.ones((3, 2), dtype= np.int8)

rng = np.random.default_rng()
for i in range(5):
    for j in range(3):
        matrix_1[i, j] = rng.integers(low=-10, high=10)
        
for i in range(3):
    for j in range(2):
        matrix_2[i, j] = rng.integers(low=-10, high=10)
        
matrix_1 @ matrix_2
````
