8. Выведите минимальное и максимальное представимое значение для каждого скалярного типа numpy

````python
def print_min_max(dtype):
    if np.issubdtype(dtype, np.integer):
        info = np.iinfo(dtype)
    elif np.issubdtype(dtype, np.floating):
        info = np.finfo(dtype)
    else:
        return f"Тип данных {dtype} не поддерживается."
    
    return f"Тип данных: {dtype}, Минимальное значение: {info.min}, Максимальное значение: {info.max}."

dtypes = [np.int8, np.int16, np.int32, np.int64, 
          np.uint8, np.uint16, np.uint32, np.uint64,
          np.float16, np.float32, np.float64]

for dtype in dtypes:
    print(print_min_max(dtype))
````
